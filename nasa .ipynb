```python
### importing 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np 
import plotly.express as px
import math 

%matplotlib inline
```

```python
df=pd.read_csv("GlobalLandTemperaturesByCity.csv")
```

```python
df.head(30)
```

```python
df.info()
```

```python
df=df.dropna()
```

```python
df.dt=pd.to_datetime(df.dt)
```

```python
df["year"]=df["dt"].dt.year
```

```python
df=df.query("year >= 2000")
```

```python
df.info()
```

```python
def func (element):
    if element[-1]=="N" or element[-1]=="E":
        element=float(element[:-1])
    elif element[-1]=="S" or element[-1]=="E":
        element=-1*float(element[:-1])
    return element 
    
df.Latitude=df.Latitude.apply(lambda x : func(x))
df.Longitude=df.Longitude.apply(lambda x : func(x))
```

```python
df.to_csv("tempertaure.csv")
```

```python
YEAR=int(input("please inter year: "))
df_show=df.query(f"year=={YEAR}")
fig = px.density_mapbox(df_show, lat='Latitude', lon='Longitude',z="AverageTemperature", radius=10,
                        center=dict(lat=0, lon=180), zoom=0,
                        mapbox_style="stamen-terrain",animation_frame="year")
fig.show()
```
